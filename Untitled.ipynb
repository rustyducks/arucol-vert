```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
timesteps = []

with open("build/poses.csv", 'r') as f:
    for l in f:
        p = l.strip().split(';')
        #d = list(map(lambda t: t.strip().split(','),p))
        #d = list(map(lambda t: list(map(float, t.strip().split(','))), p))
        d = [list(map(float, t.strip().split(','))) if t != "" else None for t in p]
        timesteps.append(d);
        
timesteps
```

```python


colors = {1.0:"red", 60.0: "blue", 61.0: "green", 62.0: "yellow", 63.0: "orange"}

# Create plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

for t in timesteps:
    if t[0] is None:
        continue
    for p in t:
        group = p[0]
        color = colors[group]
        x, y = p[1], p[2]
        ax.scatter(x, y, alpha=0.8, c=color, s=30, label=group)

plt.title('Matplot scatter plot')
plt.show()
```

```python
# Create data
N = 60
g1 = (0.6 + 0.6 * np.random.rand(N), np.random.rand(N))
g2 = (0.4+0.3 * np.random.rand(N), 0.5*np.random.rand(N))
g3 = (0.3*np.random.rand(N),0.3*np.random.rand(N))

data = (g1, g2, g3)
colors = ("red", "green", "blue")
groups = ("coffee", "tea", "water")

# Create plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

for data, color, group in zip(data, colors, groups):
    x, y = data
    ax.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

plt.title('Matplot scatter plot')
plt.legend(loc=2)
plt.show()
```
